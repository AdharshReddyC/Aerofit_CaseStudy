```python
import pandas as pd

# Creating a two-way contingency table for KP281
contingency_table_KP281 = pd.crosstab(df['Gender'], df['Product'] == 'KP281', margins=True, margins_name='Total')

# Displaying the contingency table
print("Contingency Table for KP281:")
print(contingency_table_KP281)
print()

# Computing conditional probabilities for KP281
conditional_prob_KP281 = contingency_table_KP281.div(contingency_table_KP281['Total'], axis=0)
print("Conditional Probabilities for KP281:")
print(conditional_prob_KP281)
print()

# Computing marginal probabilities for KP281
marginal_prob_KP281 = contingency_table_KP281.div(contingency_table_KP281.loc['Total'], axis=1)
print("Marginal Probabilities for KP281:")
print(marginal_prob_KP281)
```

**Conditional Probabilities for KP281:**
- **Female, True (purchasing KP281):** Among female customers, there is a 52.63% probability of purchasing the KP281 treadmill, indicating a relatively higher likelihood.
- **Male, True (purchasing KP281):** Among male customers, the probability of purchasing the KP281 treadmill is 38.46%, suggesting a comparatively lower likelihood.

**Marginal Probabilities:**
- **Female, True:** Overall, 50% of customers who purchased the KP281 treadmill are females, providing a balanced gender distribution.
- **Male, True:** Similarly, 50% of customers who purchased the KP281 treadmill are males, contributing to a well-distributed customer base.

**Insights:**
- Females show a higher probability of choosing the KP281 treadmill (52.63%) compared to males (38.46%), emphasizing potential gender-specific preferences.
- The overall balanced distribution (50%) indicates that the KP281 treadmill appeals to both genders.
